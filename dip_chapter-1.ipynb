```python
import cv2
import matplotlib.pyplot as plt

# Load an image (change 'my_image.jpg' to your actual image file)
image = cv2.imread('salman.jpg')

# Check if image was loaded
if image is None:
    print("Image not found. Please check the file name and location.")
else:
    # Convert BGR to RGB for correct color display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the image
    plt.imshow(image_rgb)
    plt.title('Loaded Image')
    plt.axis('off')
    plt.show()
```

```python
resized = cv2.resize(image, (300, 300))  # width=300, height=300
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title("Resized Image")
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Image resizing


img = cv2.imread('salman.jpg', 0)

img512 = img[::2, ::2];
img256 = img512[::2, ::2];
img128 = img256[::2, ::2];
img64 = img128[::2, ::2];
img32 = img64[::2, ::2];
            
# create figure
fig, axs = plt.subplots(2, 3, figsize=(10, 7))
  
# showing images
axs[0,0].imshow(img,cmap = 'gray')
axs[0,0].axis('off')
axs[0,0].set_title('Original image')

axs[0,1].imshow(img512, cmap = 'gray')
axs[0,1].axis('off')
axs[0,1].set_title('result512')

axs[0,2].imshow(img256,cmap = 'gray')
axs[0,2].axis('off')
axs[0,2].set_title('result256')

axs[1,0].imshow(img128,cmap = 'gray')
axs[1,0].axis('off')
axs[1,0].set_title('result128')

axs[1,1].imshow(img64,cmap = 'gray')
axs[1,1].axis('off')
axs[1,1].set_title('result64')

axs[1,2].imshow(img32,cmap = 'gray')
axs[1,2].axis('off')
axs[1,2].set_title('result32')

plt.tight_layout();
plt.show()
```
