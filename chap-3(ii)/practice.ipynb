```python
# spatial filtering
```

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Averaging filter ( box filter ).
```

```python
img1=cv2.imread("moon.png",0)
plt.imshow(img1,cmap="gray")
plt.title("original image")
plt.axis("off")
plt.show()
```

```python
mean_kernel = np.ones((3,5)) / 9
mean_filtered = cv2.filter2D(img1, -1, mean_kernel)

plt.imshow(mean_filtered, cmap='gray')
plt.title('Mean Filtered')
plt.axis('off')
plt.show()
```

```python
# Gaussian filter
```

```python
gaussian_filtered = cv2.GaussianBlur(img1, (5,5), 1)

plt.imshow(gaussian_filtered, cmap='gray')
plt.title('Gaussian Filtered')
plt.axis('off')
plt.show()
```

```python
# median filter for salt and pepper  noice reduction
```

```python
mg2=cv2.imread("noicy.png",0)

plt.imshow(img2, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.show()
```

```python
img2=cv2.imread("noicy.png",0)
median_filtered = cv2.medianBlur((img2), 3)

plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filtered')
plt.axis('off')
plt.show()
```

```python
# Laplacian filter
```

```python
laplacian = cv2.Laplacian(img1, cv2.CV_64F)
laplacian = np.uint8(np.absolute(laplacian))

plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian Sharpening')
plt.axis('off')
plt.show()
```

```python
# ist order derivative sobber filter
```

```python
sobelx = cv2.Sobel(img1, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img1, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobelx, sobely)

plt.imshow(np.uint8(sobel), cmap='gray')
plt.title('Sobel Filter')
plt.axis('off')
plt.show()
```

```python
# 3. High-Boost Filtering
```

```python
A = 1.5  # amplification factor
blurred = cv2.GaussianBlur(img2, (5,5), 0)
high_boost = cv2.addWeighted(img2, A, blurred, -1, 0)

plt.imshow(high_boost, cmap='gray')
plt.title('High-Boost Filtered')
plt.axis('off')
plt.show()
```

```python
# Band pass filter
```

```python
g1 = cv2.GaussianBlur(img1, (3,3), 1)
g2 = cv2.GaussianBlur(img1, (9,9), 2)
band_pass = cv2.subtract(g1, g2)

plt.imshow(band_pass, cmap='gray')
plt.title('Band-Pass Filter')
plt.axis('off')
plt.show()
```
